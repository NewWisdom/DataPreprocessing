```python
#Import stuff

import numpy as np
import random
import itertools
import librosa
import IPython.display as ipd
import matplotlib.pyplot as plt
import glob

%matplotlib inline
```

```python
def load_audio_file(file_path):
    input_length = 16000
    data = librosa.core.load(file_path)[0] #, sr=16000
    if len(data)>input_length:
        data = data[:input_length]
    else:
        data = np.pad(data, (0, max(0, input_length - len(data))), "constant")
    return data
def plot_time_series(data):
    fig = plt.figure(figsize=(14, 8))
    plt.title('Raw wave ')
    plt.ylabel('Amplitude')
    plt.plot(np.linspace(0, 1, len(data)), data)
    plt.show()
```

```python
folder_list = glob.glob("data/*")
print(folder_list)
print(len(folder_list))
file_list=[]

for i in range(len(folder_list)):
    file=glob.glob(folder_list[i]+"/*")
    file_list.append(file)

    index=0
    for j in file_list[i]:
        data = load_audio_file(j)
        # Adding white noise 
        wn = np.random.randn(len(data))
        data_wn = data + 0.005*wn
        plot_time_series(data_wn)
        # We limited the amplitude of the noise so we can still hear the word even with the noise, 
        #which is the objective
        ipd.Audio(data_wn, rate=16000)
        break
    break
    
    
```
